```python
# eventually load vegedict from text file

# iterate over vegedict to create: dict of occurences
counts_dict = {}
for veglist in vege_dict.values():
    for veg in veglist:
        if veg not in counts_dict:
            counts_dict[veg] = 1
        else:
            counts_dict[veg] += 1
            
# save the dict to a file

# each time run, append new data to file
```

```python
counts_list = []
for item in counts_dict.items():
    counts_list.append(item)

counts_list.sort(key=lambda x: x[0])
```

```python
# for elem in counts_list:
#     print(elem)
vegs = [x[0] for x in counts_list]
counts = [x[1] for x in counts_list]
```

```python
fig,ax = plt.subplots(figsize=(18,6))

ax.bar([2*x for x in range(len(counts))],counts,tick_label=vegs,color='orange',width=1)
ax.set_title("Running vegetable consumption in Fortman-Huft household",fontsize=18)
ax.set_ylabel('Number of weeks vegetable used',fontsize=12)
plt.xticks(rotation=80)
plt.show()
```
